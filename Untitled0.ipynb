## Datos

En primer lugar, cargamos los archivos

### Carga de datos

```r
datos_personas <- read.csv("esudb19p.csv")
datos_hogares <- read.csv("esudb19h.csv")

datos_personas$ID <- datos_personas$PB030 %/% 100
datos_hogares$ID <- datos_hogares$HB030

ecv19 <- merge(datos_personas, datos_hogares, by = "ID")
```

### Manipulación de variables

```r
ecv19 <- ecv19 %>% rename(mesNacimiento = PB130,
                          rentaBruta = PY010G,
                          rentaNeta = PY010N,
                          sexo = PB150,
                          minutosCuestionario = PB120,
                          estadoCivil = PB190)

ecv19$sexo_R <- ifelse(ecv19$sexo == 1, 0,
                       ifelse(ecv19$sexo == 2, 1, NA))

ecv19$sexo <- ifelse(ecv19$sexo == 1, "Hombre",
                     ifelse(ecv19$sexo == 2, "Mujer", NA))

ecv19$estadoCivil <-  factor(ecv19$estadoCivil)
```

## Análisis grafico


El análisis gráfico estará compuesto por:

* Histograma
* Box-plot
* Diagrama de barras
* Diagrama de dispersión


### Histograma


```r
ggplot(data = ecv19)+
  geom_histogram(aes(x = mesNacimiento), bins = 12, fill = "red", color = "black")+
  scale_x_continuous(breaks = c(1:12))+
  scale_y_continuous(limits = c(0,3500), breaks = seq(0,3500,500))+
  theme(panel.background = element_blank(), axis.line.x = element_line(), axis.line.y = element_line(),
        plot.title = element_text(hjust = 0.5, family = "Arial", size = 15), axis.text = element_text(colour = "black"))+
  labs(title = "Mes de Nacimiento ECV19", x = "Mes Nacimiento", y = "Frecuencia")+
  stat_bin(aes(x = mesNacimiento, y=..count.., label=..count..), geom="text", bins = 12, vjust = -1)
```

### Box-plot

```r
ggplot(data = subset(ecv19, minutosCuestionario <= 20))+
  geom_boxplot(aes(x = minutosCuestionario, y = factor(sexo), fill = factor(sexo)))+
  coord_flip()+
  scale_x_continuous(limits = c(0,20), breaks = seq(0,20,2.5))+
  theme_bw()+
  labs(title = "Minutos Cuestionario por Sexo", x = "Minutos cuestionario", y = "Sexo", fill = "Sexo")
```


### Diagrama de barras

```r
ggplot(data = ecv19)+
  geom_bar(aes(x = estadoCivil), fill = "green3", colour = "black")+
  scale_x_discrete(breaks = c("1","2","3","4","5"),
                   labels = c("Casado/a", "Soltero/a", "Viudo/a", "Separado/a", "Divorciado/a"), na.translate = FALSE)+
  scale_y_continuous(limits = c(0,20000))+
  theme(panel.background = element_blank(), axis.line.x = element_line(), axis.line.y = element_line(),
        plot.title = element_text(hjust = 0.5, family = "Arial", size = 15), axis.text = element_text(colour = "black"))+
  labs(title = "Estado Civil ECV19", x = "Estado Civil", y = "Frecuencia")
```

### Diagrama de dispersión

```r
options(scipen = 999)

ggplot(data = ecv19)+
  geom_point(aes(x = rentaBruta, y = rentaNeta, colour = factor(sexo)))+
  scale_y_continuous(limits = c(0,300000), breaks = seq(0,300000,50000))+
  theme(panel.background = element_blank(), axis.line.x = element_line(), axis.line.y = element_line(),
        plot.title = element_text(hjust = 0.5, family = "Arial", size = 15), axis.text = element_text(colour = "black"))+
  labs(title = "Renta Bruta vs Renta Neta", x = "Renta Bruta", y = "Renta Neta")+
  scale_colour_manual(values = c("red","black"), name = "Sexo", labels = c("Hombre", "Mujer"))+
  geom_smooth(aes(x = rentaBruta, y = rentaNeta), method = "lm")
```

## Regresión lineal

Ahora, vamos a estimar el siguiente modelo:

$$rentaNeta_i = \beta_1 + \beta_2*Mujer_i + e_i $$ $$ N(0,\theta^2)$$

```r
regresion <- lm(rentaNeta ~ sexo_R, data = ecv19)
summary(regresion)
plot(regresion)
```

**Explicación del modelo**: en media, las mujeres tienen una renta neta inferior a los hombres, al ser $\beta_2$ es  negativo y el p-valor es inferior a los niveles de significación habituales ($p-value < 0.05$).
